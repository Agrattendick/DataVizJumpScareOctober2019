```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter
```

```python
# Import jump scare csv into dataframe
js_df = pd.read_csv('Resources/jumpScare.csv')
js_df.head()
```

```python
# Scatter plot of Jump Scare Rating v Imdb rating

# Plot the data
js_df.plot.scatter(x='Jump Scare Rating', y='Imdb', figsize=(18,12))

# Grid & axis value
plt.ylim(0, 10)



plt.tight_layout()
```

```python
# Scatter plot of Jump Count v Imdb

# Plot the data
js_df.plot.scatter(x='Jump Count', y='Imdb', figsize=(18,12))

# Grid & axis value
plt.ylim(0, 10)

plt.tight_layout()
```

```python
# Scatter plot of Jump Count v Jump Scare Rating

# Plot the data
js_df.plot.scatter(x='Jump Count', y='Jump Scare Rating', figsize=(18, 12))

plt.tight_layout()
```

```python
x = Counter(js_df['Year'])
year_count_df = pd.DataFrame.from_dict(x, orient='index').reset_index()
year_count_df = year_count_df.rename(columns = {'index' : 'year', 0 : 'count'})
year_count_df = year_count_df.sort_values(by=['year'])
year_count_df = year_count_df.reset_index()
year_count_df = year_count_df.drop(columns=['index'])
year_count_df.tail()
```

```python
# Bar chart of jump scare movies by year

year_count_df.plot.bar(x='year', y='count', figsize=(18, 12))

#plt.bar(year_count_df['year'], year_count_df['count'], figsize=(18,12))

plt.tight_layout()
```

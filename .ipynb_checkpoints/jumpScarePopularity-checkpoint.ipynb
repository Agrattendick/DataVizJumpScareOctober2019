```python
# Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.linear_model import Ridge

plt.rcParams.update({'font.size':18})
plt.style.use(['dark_background'])
plt.rcParams.update({'text.color': '#615843'})
```

```python
# Import csv files

# Import Jump Scare data into dataframe
js_working_df = pd.read_csv('Resources/jumpScare.csv')

# Import all movies produced into dataframe
am_df = pd.read_csv('Resources/allMovies.csv')
```

```python
# Collect number of Jump Scare movies by year into dataframe

x = Counter(js_working_df['Year'])
js_df = pd.DataFrame.from_dict(x, orient='index').reset_index()
js_df = js_df.rename(columns = {'index' : 'Year', 0 : 'Count'})
js_df = js_df.sort_values(by=['Year'])
js_df = js_df.reset_index()
js_df = js_df.drop(columns=['index'])
```

```python
# Datawork for All Movies Dataframe

# Change datatype to int for All Movies Dataframe

am_df = am_df.astype(int)
am_df = am_df.sort_values(by=['Year'])
```

```python
# Join the dataframes, drop years (before 1973 & 2019), & divide number of jump scare movies by total movies

# Join
movies_df = js_df.join(am_df.set_index('Year'), on='Year', lsuffix='_jump', rsuffix='_total')

# Calculate percentages
movies_df['Percent_jump'] = (movies_df['Count_jump']/movies_df['Count_total'])*100

# Drop years
movies_df = movies_df[movies_df.Year != 2019]
movies_df = movies_df[movies_df.Year >= 1973]
movies_df.head()
```

```python
# Calculate regression
lr = Ridge()

lr.fit(movies_df[['Year']], movies_df['Percent_jump'])
```

```python
# Bar chart showing percent of movies that are jump scare movies
fig = plt.figure(figsize=(18, 12))
ax = fig.add_subplot(111)

# Plot the data
# Bar plot
p1 = plt.bar(movies_df['Year'], movies_df['Percent_jump'], color = '#5C131B')

# Regression line
p2 = plt.plot(movies_df['Year'], lr.coef_*movies_df['Year']+lr.intercept_, color='#1A1A13', linewidth = 4)

# Ticks
plt.xticks(movies_df['Year'], rotation = 75)
plt.xlim(1972, 2019)

# Colors
color = '#615843'
ax.spines['bottom'].set_color(color)
ax.spines['top'].set_color(color)
ax.spines['left'].set_color(color)
ax.spines['right'].set_color(color)
ax.tick_params(axis='x', colors= color)
ax.tick_params(axis='y', colors = color)

# Trendline equation
x = movies_df.Year
y = movies_df.Percent_jump
z = np.polyfit(x, y, 1)
trend = ("y=%.3fx+(%.3f)"%(z[0],z[1]))

# Titles
plt.title('Popularity of "Jump Scare" Movies Over Time', fontsize = 50, color = color)
plt.xlabel('Year of Release', fontsize = 30, color = color)
plt.ylabel('Percent of Movies labeled "Jump Scare"', fontsize = 30, color = color)
plt.legend((p1[0], p2[0]), ('"Jump Scare" Movies', 'Linear Regression\n' + trend), edgecolor = color)

# Save Figure
plt.tight_layout()
#plt.savefig('Images/Jump_Scare.png')
```
